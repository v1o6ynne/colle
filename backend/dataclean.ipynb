```python
import os
import json

def remove_empty_objects(obj):
    if isinstance(obj, dict):
        cleaned_dict = {}
        for k, v in obj.items():
            cleaned_v = remove_empty_objects(v)
            if cleaned_v is not None:
                cleaned_dict[k] = cleaned_v
        return cleaned_dict if cleaned_dict else None
    elif isinstance(obj, list):
        cleaned_list = []
        for elem in obj:
            cleaned_elem = remove_empty_objects(elem)
            if cleaned_elem is not None:
                cleaned_list.append(cleaned_elem)
        return cleaned_list if cleaned_list else None
    else:
        return obj

def process_metadata_files(folder_path):
    for filename in os.listdir(folder_path):
        if filename.endswith('.json'):
            file_path = os.path.join(folder_path, filename)
            
            # Load the metadata from the JSON file
            with open(file_path, 'r', encoding='utf-8') as file:
                metadata = json.load(file)
            
            # Remove empty objects from the metadata
            cleaned_metadata = remove_empty_objects(metadata)
            
            # Retrieve the questions
            for item in cleaned_metadata['dataset']:
                questions = item.get('questions', [])
                for question in questions:
                    print(question)
            
            # Save the cleaned metadata back to the same file
            with open(file_path, 'w', encoding='utf-8') as file:
                json.dump(cleaned_metadata, file, indent=2)
            
            print(f"Processed: {filename}")

# Specify the folder path containing the metadata files
folder_path = 'doc'

# Process all metadata files in the specified folder
process_metadata_files(folder_path)
```
